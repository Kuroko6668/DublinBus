```python
import time
import datetime
import pandas as pd
from dateutil import parser
```

```python
weatherRaw = pd.read_csv("C:/Users/35383/Documents/GTFSAPItest/GTFSR/GTFSdata/historicalWeatherDataDublin2018.csv", delimiter=",")
```

```python
#funtion to remove UTC formatting from dt_iso column

formatted_datetimes = {
    
}
def change_datetime_format(df):
    
    for datetime in df.dt_iso:
        try:      
            formatted_datetimes.update({datetime: datetime.split("+",1)[0]})
        except:
            #if not in same format or missing etc
            print('datetime ' + datetime + " has a different format")
      
        
```

```python
change_datetime_format(weatherRaw)
```

```python
#map (replace) unformatted datetimes to formatted datetimes
weatherRaw.dt_iso = pd.to_datetime(weatherRaw.dt_iso.map(formatted_datetimes))
```

```python
#drop nan features if threshold exceeded (set to 80%)
def drop_nan_column(df):
    df.dropna(axis = 1, thresh = (len(df)//100) * 80 , inplace = True)
    
drop_nan_column(weatherRaw)
```

```python
#drop features with constant values
def drop_constant_column(df):
    for column in df.columns:
        if len(df[column].unique()) == 1:
            df.drop(column,axis=1, inplace = True)
```

```python
drop_constant_column(weatherRaw)
```

```python
#drop columns that wont be used in model
weatherRaw.drop(["feels_like","temp_min","temp_max","wind_deg","weather_icon"], axis=1, inplace=True)
```
