```python
import pandas as pd
```

```python
stopsDF = pd.read_csv("stops.txt")
```

```python
#function to return number after char 'B' in the stop_id column
def substring_after(s):
       
    return s.partition('B')[2]
    
```

```python
#check if all stop_id are in same format, split by char 'B'
stoppoint_ids = []

for stop_id in stopsDF['stop_id']:
    try:       
        stoppoint_ids.append(int(substring_after(stop_id)))
    except:
        #if not in same format
        print('stop id ' + stop_id + " has a different format")
        
    
```

We can see that there are many stops with an alternative format so a different approach is needed. This data shows that there are stops that are split by letters other than 'B', are stops where there are 'BB' letter chars, and are stops where there are no letter chars e.g. 8440352291 

These stops need to be accounted for when deriving the stop number

```python
#stop numbers are also in stop_name

def reverse_substring(s):
    s = s.strip()    
    return s.split()[-1]

```

```python
#map stop nums to stop names and add to dataframe
stopsDF['stop_num'] = stopsDF['stop_name'].map(stoppoint_ids)
```
